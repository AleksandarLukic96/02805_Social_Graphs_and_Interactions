**Ex 3**: Using `def`, define a function that takes as input a number $x$ and outputs the number multiplied by itself plus three $f(x) = x(x+3)$. 

```python
def f(x):
    return x*(x+3)
```

```python
f(2)
# f(2) = 2*(2+3) = 2*5 = 10 
```

**Ex 4**: Apply this function to every element of the list `b` using a `for` loop. 

_Keep in mind, that once the list is updated, the original values in b are lost!_

```python
for i in range(len(b)):
    b[i] = f(b[i])

print(b)
```

- **JSON** (_JavaScript Object Notation_) is a lightweight data-interchange format. It is easy for humans to read and write. It is easy for machines to parse and generate.

- **XML** (_Extensible Markup Language_) also is a format for storing and transfering data. The syntax for XML is very similar to HTML, though XML is case sensitive and HTML is not.

##### Pros & Cons
- JSON is logically similar to other commonly used datatypes such as dictionaries and lists, making it much easier to read and write. A JSON file is much more compact and memory efficient as opposed to XML. 

- XML supports more datatypes than JSON and is more rigid in it's syntax - for better and for worse. This means that XML is better choice when working with many datatypes, while JSON excels in fewer datatypes. 

##### Conclusion:
Overall for the purpose of this course, JSON is far supperior to XML. This statment is supported by [amazon](https://aws.amazon.com/compare/the-difference-between-json-xml/).
